```python
from PIL import Image
from diffusers import StableDiffusionPipeline
import torch
```

```python
# Simulation de l'entrée utilisateur pour la description
descriptions = [
    "Un paysage de montagnes enneigées",
    "Une plage tropicale au coucher du soleil",
    "Une ville moderne avec des gratte-ciels illuminés"
]
```

```python
# Fonction pour générer l'image à partir de la description
def generate_image_from_description(description):
    try:
        # Charger le modèle StableDiffusionPipeline
        pipe = StableDiffusionPipeline.from_pretrained("hf-internal-testing/tiny-stable-diffusion-torch")

        # Générer l'image à partir de la description
        image = pipe(description).images[0]

        # Enregistrer l'image générée
        image.save("generated_image.png")

        # Afficher un message de confirmation
        print("Image générée avec succès.")

        # Afficher l'image générée (optionnel)
        display(Image.open("generated_image.png"))

    except Exception as e:
        print(f"Une erreur s'est produite : {str(e)}")
```

```python
# Appel de la fonction pour générer l'image avec la description donnée
# Tester chaque description
for desc in descriptions:
    generate_image_from_description(desc)
    print("=" * 50)
```
